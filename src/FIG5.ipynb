```julia
# Cell-level stats
dfcell = combine(groupby(df, :cell),
    :z => first => :z0,
    :y => (x -> x[end] - x[1]) => :ϕ,
    :time => (x -> x[end] - x[1]) => :τ,
    :x1 => mean => :x1,
    :x2 => mean => :x2,
)
dfcell.λ .= dfcell.ϕ ./ dfcell.τ
dfcell.isphys .=  isphysical(df)
d = dfcell
α = -cov(d.ϕ, d.z0) / var(d.z0)
```

```julia
data= CSV.read("./output/ratio_data_cells.csv",DataFrame);
data
arange = data.a |> unique |> sort;
brange = data.b |> unique |> sort;
```

```julia


fig,axs = subplots(figsize=(3, 3),sharex=true)

ax = axs #[0]
j = 0


slopes = []
for a in arange
    df = data[data.a .== a,:]
    j += 1
    α = [-cov(d.ϕ, d.z0) / var(d.z0) for d in groupby(df, :b)]
    αλ = [-cov(d.ϕ ./ d.τ, d.z0) / var(d.z0) * mean(d.τ) for d in groupby(df, :b)]
    push!(slopes,cov(brange,a*αλ ./α)/var(brange))
    ax.plot(brange,a*αλ ./α ,"C$(j)-",label="\$a = $(a)\$")
end
ax.legend(bbox_to_anchor=(1, 1),frameon=false)
ax.plot(brange,brange,"k--")
ax.set_ylabel("\$a\\alpha_{\\lambda}/\\alpha \$")
#ax.set_ylim([-0.5,0.5])
ax.set_xlabel("b")
# ax.set_ylabel("\$\\Sigma_{\\lambda,\\lambda}^{(1)} \$")
savefig(joinpath(FIG_PATH, "fig5_alphalam.pdf"), bbox_inches="tight", pad_inches=0.1, format="pdf")
fig
```
